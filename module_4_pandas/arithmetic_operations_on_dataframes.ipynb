# Learning Objectives:

1. Arithmetics on two dataframes
2. Arithmetics on two columns or series
3. .add() method
4. Add with fill value

6. Cummulative Sum & Product along the axis

```python
import numpy as np
import pandas as pd
```

```python
sem1_data = {"Student": ["Tom", "Katey", "Mak", "Bill", "Tom", "Rahul"],
               "Age": np.random.randint(20, 25, size=6),
               "Sub1_Score": np.random.randint(40, 100, size=6),
               "Sub2_Score": np.random.randint(40, 100, size=6),
               "Sub3_Score": np.random.randint(40, 100, size=6)
              }

sem1_df = pd.DataFrame(sem1_data)
sem1_df
```

```python
sem2_scores_data = {"Student": ["Tom", "Katey", "Mak", "Bill", "Tom", "Rahul", "Luke"],
               "Age": list(sem1_data['Age']) + [22],
               "Sub1_Score": np.random.randint(40, 100, size=7),
               "Sub2_Score": np.random.randint(40, 100, size=7),
               "Sub4_Score": np.random.randint(40, 100, size=7)
              }

sem2_df = pd.DataFrame(sem2_scores_data)
sem2_df
```

## Arithmetic operations on two dataframes

```python
sem1_df + sem2_df
```

```python
np.nan + 5
```

## Arithemtics on two columns / series

```python
sem1_df['Sub1_Score'] + sem2_df['Sub1_Score']
```

```python
sem1_df.add(sem2_df)
```

## Addition with fill value

```python
sem1_df['Sub1_Score'].add(sem2_df['Sub4_Score'], fill_value=0)
```

## Other operations

```python
sem2_df['total_score'] = 2*sem1_df['Sub2_Score'] + 0.5*sem2_df['Sub2_Score'] + sem2_df['Sub4_Score']
sem2_df
```

```python
sem2_df['percentage'] = (sem2_df['total_score']/300) * 100
sem2_df
```

## Cummulative Sum and Cummulative product

```python
sem1_df['Sub1_Score'].cumsum()
```

```python
sem1_df['Sub1_Score'].cumprod()
```

```python
sem1_df.iloc[:, 2:].cumsum(axis=1)
```
