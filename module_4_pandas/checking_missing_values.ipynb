# Learning Objectives

1. Checking for null values.
2. Dropping null values from the dataframe.
3. Filling null values in a dataframe.

```python
import numpy as np
import pandas as pd
```

```python
scores_data = {"Student": ["Tom", "Katey", "Mak", "Bill", "Tom", "Rahul"],
               "Age": np.random.randint(20, 25, size=6),
               "Sub1_Score": np.random.randint(40, 100, size=6),
               "Sub2_Score": np.random.randint(40, 100, size=6),
               "Sub3_Score": np.random.randint(40, 100, size=6)
              }

scores_df = pd.DataFrame(scores_data)
scores_df
```

```python
scores_df['Sub4_Score'] = [67, 56, 89, np.nan, 78, None]
scores_df
```

## Checking for null values

```python
scores_df.isnull()
```

```python
scores_df.isna()
```

```python
scores_df.info()
```

```python
scores_df['Sub4_Score'].isnull().sum()
```

```python
scores_df[scores_df['Sub4_Score'].isnull()].sample()
```

## Dropping null values from the dataframe

```python
cleaned_df = scores_df.dropna(axis=0)
cleaned_df
```

## Filling null values in a dataframe

```python
filled_df = scores_df.fillna(0)
filled_df
```

```python
filled_df2 = scores_df.fillna(scores_df['Sub4_Score'].median())
filled_df2
```
