```python
import numpy as np
import pandas as pd
```

```python
scores_data = {"Student": ["Tom", "Katey", "Mak", "Bill", "Rahul"],
               "Age": np.random.randint(20, 25, size=5),
               "Sub1_Score": np.random.randint(40, 100, size=5),
               "Sub2_Score": np.random.randint(40, 100, size=5),
               "Sub3_Score": np.random.randint(40, 100, size=5)
              }

left_df = pd.DataFrame(scores_data)
left_df
```

```python
right_data = {"Student": ["Tom", "Katey", "Mak"],
               "Age": [24, 20, 25],
               "Sub4_Score": np.random.randint(40, 100, size=3),
               "Sub5_Score": np.random.randint(40, 100, size=3),
              }

right_df = pd.DataFrame(right_data)
right_df
```

## Inner join

```python
merged_df = pd.merge(left_df, right_df)
merged_df
```

```python
merged_df.shape
```

```python
left_df.shape
```

```python
right_df.shape
```

## Outer joins

```python
merged_df = pd.merge(left_df, right_df, how="outer")
merged_df
```

```python
merged_df = pd.merge(left_df, right_df, how="outer", on=['Student', 'Age'])
```

```python
merged_df
```

## Left join

```python
merged_df = pd.merge(left_df, right_df, how="left")
merged_df
```

## Right join 

```python
merged_df = pd.merge(left_df, right_df, how="right")
merged_df
```

## `.join()` for merging

```python
left_df.join(right_df, lsuffix="_left", rsuffix="_right")
```

```python
right_df.join(left_df, rsuffix="_left")
```

## `concat()` functions

```python
pd.concat([left_df, right_df],  ignore_index=True)
```

```python
pd.concat([left_df, right_df], join="outer")
```
