# Learning Objectives:

1. Sorting the rows in a dataframe
2. Sorting the columns of a dataframe
3. Ranking a dataframe

```python
import numpy as np
import pandas as pd
```

```python
scores_data = {"Student": ["Tom", "Katey", "Mak", "Bill", "Tom", "Rahul"],
               "Age": np.random.randint(20, 25, size=6),
               "Sub1_Score": np.random.randint(40, 100, size=6),
               "Sub2_Score": np.random.randint(40, 100, size=6),
               "Sub3_Score": np.random.randint(40, 100, size=6)
              }

scores_df = pd.DataFrame(scores_data)
scores_df['Sub4_Score'] = [67, 56, 89, np.nan, 78, None]
scores_df
```

## Sorting the data

```python
scores_df = scores_df.sample(frac=1)
scores_df
```

```python
scores_df.sort_index()
```

```python
scores_df.sort_values(by='Age', ascending=False)
```

```python
scores_df.sort_values(by='Sub1_Score')
```

## Sorting the columns of a dataframe

```python
scores_df.sort_index(axis=1)
```

```python
scores_df.sort_index(axis=1, ascending=False)
```

```python
scores_df.sort_values(by='Student')
```

## Ranking a dataframe

```python
scores_df.rank()
```

```python
scores_df['Age'].rank()
```

```python
scores_df['Age']
```

```python
scores_df.iloc[:, 1:].rank(ascending=False, method='first')
```

```python
scores_df.iloc[:, 1:].rank(ascending=False, method='max')
```
