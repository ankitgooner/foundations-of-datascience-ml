# Learning Objectives:

1. Grouping data
2. Aggregate functions
3. Grouping by more than one column
4. Iterating over groups

```python
import numpy as np
import pandas as pd
```

```python
scores_data = {"Student": ["Tom", "Katey", "Mak", "Bill", "Katey", "Mak", "Tom"],
               "Sub1_Score": np.random.randint(40, 100, size=7),
               "Sub2_Score": np.random.randint(40, 100, size=7),
               "Sub3_Score": np.random.randint(40, 100, size=7)
              }

scores_df = pd.DataFrame(scores_data)
scores_df
```

## Grouping data

```python
scores_df.groupby("Student")
```

```python
scores_df.groupby("Student").sum()
```

## Aggregation functions

```python
scores_df.groupby("Student").mean()
```

```python
scores_df.groupby("Student").std()
```

```python
scores_df.groupby("Student").count()
```

```python
scores_df.groupby("Student").cumsum()
```

## Grouping data by more than one column

```python
scores_data = {"Student": ["Tom", "Katey", "Mak", "Bill", "Katey", "Mak", "Tom"],
               "Semester": [1,2,1,1,1,2,2],
               "Sub1_Score": np.random.randint(40, 100, size=7),
               "Sub2_Score": np.random.randint(40, 100, size=7),
               "Sub3_Score": np.random.randint(40, 100, size=7)
              }

scores_df = pd.DataFrame(scores_data)
scores_df
```

```python
scores_df.groupby([scores_df['Student'], scores_df['Semester']]).sum()
```

## Iterating over groups

```python
for name, group_data in scores_df.groupby('Student'):
    print(name.upper())
    print(group_data)
```

```python
scores_df.groupby("Student")['Sub1_Score'].mean()
```
