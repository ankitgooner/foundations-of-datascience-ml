## Learning Objectives

1. Plotting a dataframe
2. Plotting multiple columns

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = (12,8)
```

## Reading the file

```python
df = pd.read_csv("../data/MSFT.csv", index_col=0, parse_dates=True)
df.head()
```

```python
df.info()
```

## Plotting data from dataframe

```python
##initialize the plot
fig, ax = plt.subplots(1,1)

##add plot to the axis
df['Close'].plot(ax=ax)

ax.set_xlim(['2018-01-09', '2020-12-12'])

ax.set_title("Peaks and bottoms in last 5 years")
```

## Plotting multiple columns

```python
##initialize the plot
fig, ax = plt.subplots(1,1)

##add plot to the axis
df[['Open', 'Close', 'Volume']].plot(ax=ax)



```
