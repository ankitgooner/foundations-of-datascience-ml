## Learning Objectives

1. Plotting data in the subplots
2. Creating a new figure with a grid of subplots
3. Sharing X-Y axis and adjusting spacing between axes

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
numbers = np.arange(1,50,5)
numbers
```

```python
plt.plot(numbers)
plt.show()
```

```python
fig = plt.figure()
```

```python
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

```python
fig
```

```python
ax4 = fig.add_subplot(2,2,4)
```

## Plotting data in subplots

```python
ax1.plot(np.random.randn(50).cumsum())
```

```python
fig
```

```python
ax2.plot(np.random.randn(50).cumprod())
fig
```

## Creating a new figure with a grid of subplots

```python
fig, axes = plt.subplots(2,3)
```

```python
axes
```

```python
fig
```

```python
axes[0, 1].plot(np.random.randn(50).cumsum())
```

```python
fig
```

## Sharing X-Y axis and adjusting the space

```python
fig, axes = plt.subplots(2,3, sharex=True, sharey=True)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
```

```python
ax.plot(numbers)
```
