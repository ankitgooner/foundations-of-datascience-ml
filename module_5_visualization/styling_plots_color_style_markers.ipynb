## Learning Objectives

1. Plotting a chart with x & y values
2. Adding color
3. Adding line style
4. Plotting a line with a marker

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = (12,8)
```

## Plotting a chart with x & y values

```python
x = np.random.randint(10, 100, size=10)
y = 4*x + 5

```

```python
print(x)
print("Y-------")
print(y)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x,y)
```

## Adding color

```python
fig, ax = plt.subplots(1,1)
ax.plot(x,y, color='lightcoral')
```

## Adding linestyles

```python
fig, ax = plt.subplots(1,1)

ax.plot(x,y, color='lightcoral', linestyle='--')
```

## Adding markers

```python
fig, ax = plt.subplots(1,1)
ax.plot(x,y, color='lightcoral', linestyle='--', marker='o')
```
