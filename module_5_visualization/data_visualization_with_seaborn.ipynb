## Learning Objectives

1. Seaborn library
2. Plotting a relational plot of the data
3. Countplot
4. Histogram/Distplot
5. Pairplot of all the numerical columns
6. Boxplot for total_bill


```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.rcParams["figure.figsize"] = (12,8)
```

```python
sns.set_theme(style='darkgrid')
```

## Seaborn dataset

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

## Plotting a relational plot

```python
sns.relplot(x='age', y='fare', data=df)
```

## Countplot

```python
sns.countplot(df['sex'], data=df)
```

```python
df['sex'].value_counts()
```

## Age distribution using hist/distplot

```python
sns.histplot(x='age', data=df, bins=40)
```

```python
sns.distplot(df['age'])
```

## pairplot

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df)
```

## Boxplot

```python
sns.boxplot(x=tips_df['sex'], y=tips_df['tip'])
```
