## Learning Objectives

1. Setting axis labels of a plot
2. Setting ticks and ticklabels
3. Adding legends to a plot

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = (12,8)
```

## Setting axis labels to a plot

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

data = np.random.randn(1000).cumsum()

ax.plot(data)

ax.set_xlabel("Index")
ax.set_ylabel("Value")

ax.set_title("Cummulative sum of random normal data")
plt.show()
```

## Setting ticks and ticklabels

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

data = np.random.randn(1000).cumsum()

ax.plot(data)
ax.set_xlabel("Index")
ax.set_ylabel("Value")

ax.set_title("Cummulative sum of random normal data")

ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(["Start", "1st_Quarter", "2nd_Quarter", "3rd_Quarter", "Last_Quarter"],
                  rotation=45,
                  fontsize="medium")

plt.show()

```

## Adding legends to a plot

```python
x1 = np.random.randn(1000).cumsum()
x2 = np.random.randn(1000).cumsum()
x3 = np.random.randn(1000).cumsum()

fig, ax = plt.subplots(1,1)

ax.plot(x1, color='g', label='X1')
ax.plot(x2, color='r', label='X2')
ax.plot(x3, color='k', label='X3')

ax.set_xlabel("X")
ax.set_title("X1, X2, X3 for 4 quarters")

ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(["Start", "1st_Quarter", "2nd_Quarter", "3rd_Quarter", "Last_Quarter"],
                  rotation=45,
                  fontsize="medium")


ax.legend(loc='best')
plt.show()
```
