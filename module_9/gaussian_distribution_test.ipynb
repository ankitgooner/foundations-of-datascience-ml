# Gaussian Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian_dist(x, mu, sigma):
    dist = (1 / np.sqrt(2 * np.pi * sigma ** 2)) *\
    np.exp(-(1 / (2 * sigma ** 2)) * (x - mu) ** 2)
    return dist
```

```python
x_data = np.arange(-5, 5, 0.001)
y_data = gaussian_dist(x_data, mu=0.0, sigma=1.0)
```

```python
plt.plot(x_data, y_data)
```

```python
y_data1 = gaussian_dist(x_data, mu=0.0, sigma=1.0)
y_data2 = gaussian_dist(x_data, mu=1.0, sigma=0.5)
y_data3 = gaussian_dist(x_data, mu=-1.0, sigma=2.0)

plt.figure(figsize=(12,8))
plt.plot(x_data, y_data1)
plt.plot(x_data, y_data2)
plt.plot(x_data, y_data3)
plt.legend(['y_data1', 'y_data2', 'y_data3'])
```

```python
from scipy.stats import norm

mu = 0.0
sigma = 1.0
pdf = norm.pdf(x_data, mu, sigma)
plt.plot(x_data, pdf)
```

```python
##cumulative density function
from scipy.stats import norm

cdf = norm.cdf(x_data)
plt.plot(x_data, cdf)
```
