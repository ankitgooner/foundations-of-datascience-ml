# Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##simulate 50 die rolling experiment
n_rolls = np.random.randint(1, 7, 50)
n_rolls
```

```python
np.mean(n_rolls)
```

```python
##simulate the exp 1000 times
n_samples = 1000

##store the mean of each sample
means = [ np.mean(np.random.randint(1, 7, 50)) for _ in range(n_samples + 1)]
means
```

```python
plt.hist(means, bins=50)
```
