# Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm
```

```python
##simulate a 50 die-roll experiment
n_rolls = np.random.randint(1,7, 50)
n_rolls
```

```python
np.mean(n_rolls)
```

```python
##simulating 1000 trials of 50 die-roll experiments
n_trials = 1000

##storing means of each trial
means = [ np.mean(np.random.randint(1,7,50)) for _ in range(n_trials + 1) ]
means
```

```python
plt.hist(means, bins=50)
```
