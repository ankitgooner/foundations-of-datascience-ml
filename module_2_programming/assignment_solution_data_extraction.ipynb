Here is a sample of what my data looks like:


```python
import requests
```

```python
popular_req = requests.get("https://api.themoviedb.org/3/movie/popular?api_key=9e0a41a1a98bb5e6aa885d4ffcfbe688&language=en-US&page=1")

data = popular_req.json()
```

```python
data['results']
```

```python
title = []
adult_rating = []
original_language = []
popularity = []
release_date = []
vote_average = []

for movie in data['results']:
    title.append(movie['title'])
    adult_rating.append(movie['adult'])
    original_language.append(movie['original_language'])
    popularity.append(movie['popularity'])
    release_date.append(movie['release_date'])
    vote_average.append(movie['vote_average'])
    
```

```python
title
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'title': title,
    'release_date': release_date,
    'adult_rating': adult_rating,
    'original_language': original_language,
    'popularity': popularity,
    'vote_average': vote_average
})



df.to_csv("<<filename.csv>>")
```
